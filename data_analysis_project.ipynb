```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv("Diwali Sales Data.csv",encoding="unicode_escape")
print(data)
```

```python
data.shape
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
print (data.info())
```

```python
pd.isnull(data).sum()
```

```python
data.info()
```

```python
data["Amount"]=data["Amount"].astype("int")
```

```python
data["Amount"].mean()
```

```python
data.replace(np.nan,9453,inplace=True)
print(data)
```

```python
# EXPOSITORY DATA ANALYSIS
```

```python
data.columns
```

```python
# GENDEER
```

```python
ax=sns.countplot(x="Gender", data=data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_gen=data.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Gender",y="Amount" ,data=sale_gen)
plt.show()
```

```python
# From above graph we can see that the most of the buyers are "female"
```

```python
# AGE GROUP
```

```python
ax=sns.countplot(x="Age Group",data=data,hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=data.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount" ,data=sales_age)
plt.show()
```

```python
# We can see that most of the buyers are of age group between 26-35 and mostly females.
```

```python
# STATE
```

```python
state=data.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False)
sns.barplot(x="State",y="Orders" ,data=state)
sns.set(rc={'figure.figsize':(20,4)})
plt.show()
```

```python
state=data.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="State",y="Amount" ,data=state)
sns.set(rc={'figure.figsize':(20,4)})
plt.show()
```

```python
# we can see that unexpectedly most of the order are from Uttar Pardesh,Maharashtra and Kamataka
```

```python
# MARTIAL STATUS
```

```python
    data.columns
```

```python
ax=sns.countplot(x="Marital_Status",data=data)
for bars in ax.containers:
    ax.bar_label(bars)
    sns.set(rc={'figure.figsize':(7,6)})
```

```python
sale_gen=data.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Marital_Status",y="Amount" ,data=sale_gen)
sns.set(rc={'figure.figsize':(7,6)})
plt.show()
```

```python
# We can see that most of the buyers are married mostly (Females)
```
