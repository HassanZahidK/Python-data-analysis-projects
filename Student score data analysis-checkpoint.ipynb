```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

GENDER

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(x="Gender" ,data=data)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# we can see that numbers of females is more than the numbers of male
```

```python
data.columns
```

```python
gb=data.groupby(["ParentEduc"]).agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
plt.title("RElationship OF PARENT-EDUCATION WITH STUDENT SCORES")
sns.heatmap(gb ,annot=True)
plt.show()
```

```python
# we have concluded that the education of the parents have good impact on their childrens
```

MARTIAL STATUS

```python
data.columns
```

```python
gb=data.groupby(["ParentMaritalStatus"]).agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
plt.title("RElationship OF PARENT-MARTIAL STATUS WITH STUDENT SCORES")
sns.heatmap(gb,annot=True)
plt.show()
```

```python
# we have analyzed the martial status of parent have no or negligible impact on student score
```

```python
# DISTRIBUTION OF AN ATHNIC GROUP
```

```python
data.columns
```

```python
groupA=data.loc[(data["EthnicGroup"]=="group A")].count()
groupB=data.loc[(data["EthnicGroup"]=="group B")].count()
groupC=data.loc[(data["EthnicGroup"]=="group C")].count()
groupD=data.loc[(data["EthnicGroup"]=="group D")].count()
groupE=data.loc[(data["EthnicGroup"]=="group E")].count()
list=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
l=["group A","group B","group C","group D","group E"]
plt.pie(list,autopct="%1.2f%%",labels=l) 
plt.show()
```

```python
# CONCLUSION:Most of the students are female,education of the parents have good impact on student scores but marital status of the parents have no or negligible impact on students score
```
